```python
import pandas as pd
from google.colab import files

def sorted_data():

  # added code to combine the 5 dataframes, filling missing values with NA
  url="https://github.com/Nikitatamotia/Winter-Project-/blob/main/Combine_happiness%202.xlsx?raw=true"
  happiness = pd.read_excel(url)
  Denmark_happiness=happiness[happiness["Country"]=="Denmark"]
  Burundi_happiness=happiness[happiness["Country"]=="Burundi"]
  Denmark_happiness["Country"]=="Denmark"
  Burundi_happiness["Country"]=="Burundi"
  H_Denmark= pd.pivot_table(Denmark_happiness,index =['Country','year'], values=['Happiness Score','Economy (GDP per Capita)','Freedom','Trust (Government Corruption)'])
  H_Burundi= pd.pivot_table(Burundi_happiness,index =['Country','year'], values=['Happiness Score','Economy (GDP per Capita)','Freedom','Trust (Government Corruption)'])
  display (H_Denmark)
  display (H_Burundi)
  combine_whole = pd.concat([Denmark_happiness,Burundi_happiness],ignore_index = True)
  combine_whole.to_excel("highest_lowest_happiness.xlsx")
  files.download("highest_lowest_happiness.xlsx")
  
happiness=""
sorted_data()
```

```python
import pandas as pd

def sorted_correlation():

# code below to return a series sorted by highest Happiness Score
  url="https://github.com/Nikitatamotia/Winter-Project-/blob/main/Combine_happiness.xlsx?raw=true"
  happiness = pd.read_excel(url)
  HD_correlation = happiness.corr(method ="pearson", min_periods=5)["Happiness Score"].sort_values(ascending=False)
  display (HD_correlation)
  #return HD_correlation

happiness = " "
sorted_correlation()
```

It is very easy to see a strong positive correlation between the GDP per capita and the score of happiness between countries. In fact there are similar positives correlations like that with our other factor variables.

Our GDP variable has the strongest correlation with our happiness variable (0.79), after that comes life expectancy (0.74) and Freedom (0.55). The generosity felt within a country variable has the lowest correlation (0.13).
