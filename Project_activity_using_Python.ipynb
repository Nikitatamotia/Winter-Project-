```python
import pandas as pd
from google.colab import files 

def combine_whole():
  # added code to combine the 5 dataframes, filling missing values with NA
  url ="https://github.com/Nikitatamotia/Winter-Project-/blob/main/WHD.xls?raw=true"
  Happiness_2015_df = pd.read_excel(url,sheet_name = "2015")
  Happiness_2016_df = pd.read_excel(url,sheet_name = "2016")
  Happiness_2017_df = pd.read_excel(url,sheet_name = "2017")
  Happiness_2018_df = pd.read_excel(url,sheet_name = "2018")
  Happiness_2019_df = pd.read_excel(url,sheet_name = "2019")
  
  Happiness_2015_df["year"]=2015
  Happiness_2016_df["year"]=2016
  Happiness_2017_df["year"]=2017
  Happiness_2018_df["year"]=2018
  Happiness_2019_df["year"]=2019
  
  combine_whole = pd.concat([Happiness_2015_df,Happiness_2016_df,Happiness_2017_df, Happiness_2018_df, Happiness_2019_df],ignore_index = True)
  display (combine_whole)
  combine_whole.to_excel("Combine_happiness.xlsx")
  files.download("Combine_happiness.xlsx")
  return combine_whole
  
# created new dataframe from function above

combine_df = combine_whole()

#test code to check new dataframe contains null values 
actual= combine_df.isnull().values.any()
expected = True 

if actual == expected:
  print("Test passed", actual)
else:
  print("Test failed, expected", expected, "got", actual)


```

```python
import pandas as pd
from google.colab import files

def sorted_data():

  # added code to combine the 5 dataframes, filling missing values with NA
  url="https://github.com/Nikitatamotia/Winter-Project-/blob/main/Combine_happiness%202.xlsx?raw=true"
  happiness = pd.read_excel(url)
  Denmark_happiness=happiness[happiness["Country"]=="Denmark"]
  Burundi_happiness=happiness[happiness["Country"]=="Burundi"]
  Denmark_happiness["Country"]=="Denmark"
  Burundi_happiness["Country"]=="Burundi"
  H_Denmark= pd.pivot_table(Denmark_happiness,index =['Country','year'], values=['Happiness Score','Economy (GDP per Capita)','Trust (Government Corruption)'])
  H_Burundi= pd.pivot_table(Burundi_happiness,index =['Country','year'], values=['Happiness Score','Economy (GDP per Capita)','Trust (Government Corruption)'])
  display (H_Denmark)
  display (H_Burundi)
  combine_whole = pd.concat([Denmark_happiness,Burundi_happiness],ignore_index = True)
  combine_whole.to_excel("highest_lowest_happiness.xlsx")
  files.download("highest_lowest_happiness.xlsx")
  
happiness=""
sorted_data()
```
